**Step 5: Check Missing Values**

```python

df.isnull().sum()
```

**Step 6: Handle Missing Values**

```python
for col in df.select_dtypes(include=['float64','int64']).columns:
    df[col].fillna(df[col].mean(), inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

df.isnull().sum()
```

**Step 7: Remove Duplicate Records**

```python
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]
print('Duplicates removed:', before - after)
```

**Step 8: Encode Categorical Column**

```python
encoder = LabelEncoder()
df['Category_encoded'] = encoder.fit_transform(df['Category'])
df.head()
```

**Step 9: Feature Scaling**

```python
scaler = StandardScaler()
df['Revenue_per_Unit'] = df['Sales'] / df['Quantity']
cols = ['Quantity', 'Sales', 'Profit', 'Revenue_per_Unit']
df[cols] = scaler.fit_transform(df[cols])
df.head()
```

**Step 10: Save Final Dataset**

```python
df.to_csv('assignment2_cleaned_ecommerce.csv', index=False)
print('Saved: assignment2_cleaned_ecommerce.csv')
```

Moving into feature engineering, the categorical Category column was transformed into a numerical Category_encoded column using label encoding, making it suitable for machine learning models.
