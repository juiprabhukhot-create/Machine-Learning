#  Assignment-4: Support Vector Machine (SVM)

Objective: Demonstrate SVM classification using an appropriate dataset and visualize results.

Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print('Libraries imported')
```

Load Dataset

```python
iris = load_iris(as_frame=True)
df = iris.frame.copy()
df.rename(columns={'target':'class'}, inplace=True)
df['class'] = df['class'].apply(lambda i: iris.target_names[i])
df.head()
```

Explore Dataset

```python
print('Shape:', df.shape)
print('Classes:', df['class'].unique())
df.describe()
```

Select Features

```python
X = df[['petal length (cm)','petal width (cm)']]
y = df['class']
X.head()
```

Train–Test Split (70:30)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print('Training size:', X_train.shape)
print('Testing size:', X_test.shape)
```

Build SVM Model (Linear Kernel)

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
print('Model trained')
```

Predict Test Data

```python
y_pred = svm.predict(X_test)
list(y_pred)[:10]
```

Compare Predictions (Correct / Wrong)

```python
results = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
results['Status'] = results.apply(lambda r: 'Correct' if r.Actual == r.Predicted else 'Wrong', axis=1)
results.head(20)
```

Evaluate Model Performance

```python
print('Accuracy:', accuracy_score(y_test, y_pred))
print('\nConfusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('\nClassification Report:\n', classification_report(y_test, y_pred))
```

Visualize Decision Boundary

```python
x_min, x_max = X.iloc[:,0].min()-1, X.iloc[:,0].max()+1
y_min, y_max = X.iloc[:,1].min()-1, X.iloc[:,1].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))

# Get class to integer mapping
class_to_int = {name: i for i, name in enumerate(iris.target_names)}

# Prepare meshgrid input as DataFrame with feature names
mesh_input = pd.DataFrame(np.c_[xx.ravel(), yy.ravel()], columns=X.columns)

# Predict on the meshgrid and convert string labels to integers
Z_str = svm.predict(mesh_input)
Z = np.array([class_to_int[label] for label in Z_str]).reshape(xx.shape)

plt.figure(figsize=(7,5))
plt.contourf(xx, yy, Z, alpha=0.25)
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=[class_to_int[label] for label in y], cmap=plt.cm.RdYlBu)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('SVM Decision Boundary on Iris Dataset')
plt.show()
```

# **Conclusion:**
In this practical, an SVM classifier with a linear kernel was implemented on the Iris dataset to classify flower species. The model achieved good accuracy and effectively separated classes using a decision boundary. The visualization helped in understanding how SVM finds the optimal separating line, showing that it is a powerful and reliable method for classification tasks.
